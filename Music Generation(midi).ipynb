# processing all the files

```python
"""creating nots file ans saving it as notes.pkl"""
notes=[]
for file in glob.glob('midi_songs/*.mid'):
    midi=converter.parse(file)
    elements_to_parse=midi.flat.notes
    for elements in elements_to_parse:
        if isinstance(elements,note.Note):
            notes.append(str(elements.pitch))
        elif isinstance(elements,chord.Chord):
            notes.append('+'.join(str(n) for n in elements.normalOrder))

    
```

```python
print(len(notes))
```

```python
#save notes in the file notes.pkl
with open("notes.pkl",'wb') as f:
    pickle.dump(notes,f)
```

```python
#loading notes from notes.pkl
with open("notes.pkl",'rb') as f:
    notes=pickle.load(f)
```

```python
print("total notes are ",len(notes))
print("total unique notes ",len(set(notes)))
```

```python
sequence_length=100
```

```python
pitchname=sorted(set(notes))
```

```python
pitchname
```

```python
ele_to_int=dict((ele,idx) for idx,ele in enumerate(pitchname))
```

```python
ele_to_int
```

```python
network_input=[]
network_output=[]
for i in range(len(notes)-sequence_length):
    seq_in=notes[i:i+sequence_length]
    seq_out=notes[i+sequence_length]
    
    network_input.append([ele_to_int[ch] for ch in seq_in])
    network_output.append(ele_to_int[seq_out])
```

```python
print(len(network_input),network_input[0])
print(len(network_output))
```

```python
n_patterns=len(network_input)
```

```python
network_input=np.reshape(network_input,(n_patterns,sequence_length,1)) # converting it into 3d becaause lstm accepts in 3d
```

```python
network_input.shape
```

```python
note
```

```python
n_vocab=len(set(notes))
```

```python
normalised_netwrok_input=network_input/float(n_vocab)
```

```python
#encoding netwrod output classes into one hot vector
```

```python
network_output=np_utils.to_categorical((network_output))
```

```python
print(network_output.shape)
print(normalised_netwrok_input.shape)
```

# create model

```python
from keras.models import Sequential,load_model
from keras.layers import*
from keras.callbacks import EarlyStopping,ModelCheckpoint
```

```python
model=Sequential()
```

```python
model.add(LSTM(units=512,input_shape=(normalised_netwrok_input.shape[1],normalised_netwrok_input.shape[2]),return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512,return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(512))
model.add(Dropout(0.3))
model.add(Dense(n_vocab,activation='softmax'))
          
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam')
```

# predictions

```python
sequence_length=100
network_input=[]
for i in range(len(notes)-sequence_length):
    seq_in=notes[i:i+sequence_length]
    network_input.append([ele_to_int[ch] for ch in seq_in])
```

```python
start=np.random.randint(len(network_input)-1)

int_to_ele=dict((num,ele) for num,ele in enumerate(pitchname))

pattern=network_input[start]
prediction_output=[]

#predicting next 400 notes
for no in range(400):
    prediction_input=np.reshape(pattern,(1,len(pattern),1))
    prediction_input=prediction_input/float(n_vocab)
    
    prediction=model.predict(prediction_input,verbose=0)
    idx=np.argmax(prediction)
    
    result=int_to_ele[idx]
    
    prediction_output.append(result)
    
    pattern.append(idx)
    
    pattern=pattern[1:]
```

```python
# creating midi file from prediction
```

```python
offset=0   #time
output_notes=[]
for pattern in prediction_output:
    #if the pattern is the chord
    if ('+' in pattern) or pattern.isdigit():
        notes_in_chord=pattern.split('+')
        temp_notes=[]
        for current in notes_in_chord:
            new_note=note.Note(int(current))
            new_note.storedinstrument=instrument.Piano()
            temp_notes.append(new_note)
        
        new_chord=chord.Chord(temp_notes)
        new_chord.offset=offset
        output_notes.append(new_chord)
        
    #if patterns is note
    else:
        new_note=note.Note(pattern)
        new_note.offset=offset
        new_note.storedinstrument=instrument.Piano()
        output_notes.append(new_note)
    offset+=0.5
```

```python
#create stream object from the created file
midi_stream=stream.Stream(output_notes)
midi_stream.write('midi',fp='generated_midi.mid')
```
